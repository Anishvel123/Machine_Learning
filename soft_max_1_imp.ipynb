```python
def one_hot_encoding(y):
    col = np.unique(y).shape[0]
    encode = np.zeros((y.shape[0], col))
    for idx, val in enumerate(y):
        encode[idx][val] = 1
    return encode
```

```python
def softmax(z):
    return (np.exp(z).T / np.sum(np.exp(z), axis = 1)).T
```

```python
def training(X, y, y_enc, w, b):
    costs = []
    ite = []
    for itr in range(100000):
        cost, smax = loss_calc(X, y, w, b, y_enc)
        dW, db = gradiant_calc(X, smax, y_enc)
        w = w - 0.01 * dW
        b = b - 0.01 * db
        if (itr % 1000 == 0):
            print(w, b)
            costs.append(cost)
            ite.append(itr)
    return w, b, costs, ite
```

```python
def loss_calc(X, y, w, b, y_enc):
    z = X @ w + b
    smax = softmax(z)
    cost = -1 * np.sum(np.log(smax[range(y.shape[0]) , y])) / y.shape[0]
    return cost, smax
```

```python
def gradiant_calc(X, smax, y_enc):
    dW = np.dot(X.T, smax - y_enc) / X.shape[1]
    db = np.sum(smax - y_enc, axis = 0) / X.shape[1]
    return dW, db
```
