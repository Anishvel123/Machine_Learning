```python
import pandas as pd
import matplotlib.pyplot as pyp
import sklearn
from sklearn import linear_model

df = pd.read_csv('1.csv')
df
```

```python
%matplotlib inline

pyp.scatter(df.area, df.price)
pyp.xlabel('area')
pyp.ylabel('price')
```

```python
reg = linear_model.LinearRegression()
type(df[['price']])
reg.fit(df[['area']], df[['price']])
reg.predict([[3100]])
```

```python
m = reg.coef_
c = reg.intercept_

y = m * 3100 + c
y
```

```python
d = pd.read_csv("areas.csv")
d
```

```python
p = reg.predict(d[['area']])
d['prices'] = p
d
```

```python
d.to_csv("areas.csv", index = False)
```

```python
pyp.scatter(df.area, df.price,color = "orange", marker = '+')
pyp.xlabel('area')
pyp.ylabel('price')
pyp.plot(df['area'], reg.predict(df[['area']]), color = 'red')
```

```python
pyp.scatter(d['area'], d['prices'], color = 'orange', marker = '+')
pyp.xlabel('area')
pyp.ylabel('prices')
pyp.plot(d['area'], reg.predict(d[['area']]), color = 'red')
```
