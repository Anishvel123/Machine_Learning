```python
import numpy as np
import pandas as pd
import math

def gradient_descent(x, y):
    # initialize learning rate, iterations
    learning_rate = 0.0005
    iterations = 10000
    n = len(x)
    cur_m = cur_c = 0
    c = 0
    for i in range(iterations):
        # draw the predicted line
        y_predict = cur_m * x + cur_c
        s = 0
        # find the cost of the predicted line
        for j in range(n):
            s += (y[j] - y_predict[j]) ** 2
        cost = (1/n) * s
        # partial derivative of slope and intercept
        pm = -(2/n) * sum(x * (y - y_predict))
        pb = -(2/n) * sum((y - y_predict))
        # adjusting the slope and intercept
        cur_m = cur_m - learning_rate * pm
        cur_c = cur_c - learning_rate * pb
        if (math.isclose(c, cost, rel_tol = 1e-20)):
            break
        c = cost
        print("m: ", cur_m, "c: ", cur_c, "cost: ", cost)

df = pd.read_csv("test_scores.csv")
x = np.array(df.math)
y = np.array(df.cs)
print(x)
print(y)
gradient_descent(x, y)
```
