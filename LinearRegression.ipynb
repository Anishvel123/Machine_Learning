```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv("Salary_Data(2).csv")
df.head()
```

```python
plt.plot(df.iloc[:, 0], df.iloc[:, 1], 'b*')
```

```python
x = np.array(df.iloc[:, 0])
y = np.array(df.iloc[:, 1])
```

```python
def coef_calc(x, y):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    n = np.size(x)
    mxy = np.sum(y * x) - n * mean_y * mean_x
    mxx = np.sum(x * x) - n * mean_x * mean_x
    b1 = mxy / mxx
    b0 = mean_y - b1 * mean_x
    return (b0, b1)
```

```python
plt.scatter(x, y)
c, m = coef_calc(x, y)
plt.plot(x, m * x + c, 'g--')
```
