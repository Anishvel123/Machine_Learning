```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("enjoysport(1).csv")
df
```

```python
concept = np.array(df.iloc[:, 0:-1])
target = np.array(df.iloc[:, -1])
```

```python
def cea(concept, target):
    s_hyp = concept[0].copy()
    g_hyp = [['?' for i in range(len(s_hyp))] for j in range(len(s_hyp))]
    for k, h in enumerate(concept):
        if target[k] == 'yes':
            for x in range(len(h)):
                if (s_hyp[x] != h[x]):
                    s_hyp[x] = '?'
                    g_hyp[x][x] = '?'
        if target[k] == 'no':
            for y in range(len(h)):
                if (s_hyp[y] != h[y]):
                    g_hyp[y][y] = s_hyp[y]
                else:
                    g_hyp[y][y] = '?'
    return s_hyp, g_hyp
```

```python
print(cea(concept, target))
```
