```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
```

```python
digits = load_digits()
dir(digits)
```

```python
X = digits.data
y = digits.target
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c = y)
plt.title("PCA of the digits dataset")
plt.xlabel("Component 1")
plt.ylabel("Componenet 2")
plt.show()
```

```python
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.title("2d reduction of the digit dataset")
plt.show()
```

```python
linear_svc = SVC(C = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
linear_svc.fit(X_train, y_train)
```

```python
linear_svc.score(X_test, y_test)
```

```python
svc = LinearSVC(C = 0.1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
svc.fit(X_train, y_train)
```

```python
svc.score(X_test, y_test)
```

```python
poly_kernelsvc = SVC(kernel = 'poly', degree = 3, C = 1)
poly_kernelsvc.fit(X_train, y_train)
```

```python
poly_kernelsvc.score(X_test, y_test)
```
