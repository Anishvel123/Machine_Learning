```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
dataset = load_iris()
```

```python
type(dataset.data)
```

```python
df = pd.DataFrame(dataset.data)
df['target'] = dataset.target
df = df[df.target != 2]
X = df.drop('target', axis = 1)
y = df.target
X_numpy = X.to_numpy()
y_numpy = y_numpy = y.to_numpy().reshape(-1, 1)
X_numpy = np.hstack([X_numpy, np.ones((X_numpy.shape[0], 1))])
w = np.zeros((X_numpy.shape[1], 1))
y_numpy = y_numpy.reshape(len(y_numpy), 1)
y_numpy
```

```python
def training(X_numpy, y_numpy, w, iters, alfa):
    for itr in range(iters):
        y_pred = sigmoid(comp_z(X_numpy, w))
        eps = 1e-15
        y_pred = np.clip(y_pred, eps, 1 - eps)
        neg_log_likelihood = -np.mean(y_numpy * np.log(y_pred) + (1 - y_numpy) * np.log(1 - y_pred))
        grad_vector = np.dot(X_numpy.T, y_pred - y_numpy) / X_numpy.shape[0]
        w = w - alfa * grad_vector
        if (itr % 100 == 0):
            print("cost", neg_log_likelihood)
    return w
```

```python
def comp_z(X, w):
    return np.dot(X, w)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
new_w = training(X_numpy, y_numpy, w, 1000000, 0.001)
```
